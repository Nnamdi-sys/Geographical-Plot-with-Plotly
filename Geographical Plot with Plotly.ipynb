```python
import pandas as pd
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_African_countries_by_GDP_(nominal)')[0][1:].drop(columns='Notes')
df.head()
```

```python
import chart_studio.plotly as py
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
data = dict(type = 'choropleth',
            locations = df['Country'],
            locationmode = 'country names',
            #colorscale= 'Portland',
            text= df['Country'],
            z=df['Nominal GDP per capita (US$)'],
            colorbar = {'title':'Nominal GDP per capita (US$)'})
```

```python
layout = dict(title = '2019 Africa GDP Per Capita',
              geo = {'scope':'africa'})
```

```python
choromap = go.Figure(data = [data],layout = layout)
```

```python
iplot(choromap)
```
